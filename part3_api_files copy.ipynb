```python
import requests

url = "https://dummyjson.com/products?limit=20"
response = requests.get(url)

if response.status_code == 200:  # Checking if request was successful
    data = response.json()
    products = data["products"]
    print(f"Fetched {len(products)} products successfully!\n")
else:
    print(f"Failed to fetch data. Status code: {response.status_code}")




print(f"{'ID':<4} | {'Title':<30} | {'Category':<13} | {'Price':<8} | {'Rating'}") # Header
print(f"{'-'*4}-|-{'-'*30}-|-{'-'*13}-|-{'-'*8}-|-{'-'*6}")


for product in products:                    # Rows
    pid      = product["id"]
    title    = product["title"][:30]        #  truncate long titles
    category = product["category"][:13]     #  truncate long categories
    price    = f"${product['price']}"
    rating   = product["rating"]

    print(f"{pid:<4} | {title:<30} | {category:<13} | {price:<8} | {rating}")
```

```python
#From the 20 fetched products, filter only those with a rating ≥ 4.5.

top_rated = [product for product in products if product["rating"] >= 4.5]

print(f"Products with rating >= 4.5: {len(top_rated)}\n")

for product in top_rated:
    pid      = product["id"]
    title    = product["title"][:30]
    category = product["category"][:13]
    price    = f"${product['price']}"
    rating   = product["rating"]
    print(f"{pid:<4} | {title:<30} | {category:<13} | {price:<8} | {rating}")
```

```python
#Sort the filtered list by price in descending order

sorted_products = sorted(top_rated, key=lambda p: p["price"], reverse=True)

print(f"Top rated products sorted by price (high to low):\n")



for product in sorted_products:
    pid      = product["id"]
    title    = product["title"][:30]
    category = product["category"][:13]
    price    = f"${product['price']}"
    rating   = product["rating"]
    print(f"{pid:<4} | {title:<30} | {category:<13} | {price:<8} | {rating}")
```

```python
#Task 3 — Exception Handling

def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"


print(safe_divide(10, 2))       # Normal division
print(safe_divide(10, 0))       # Divide by zero
print(safe_divide("ten", 2))
```

```python
def read_file_safe(filename):
    try:
        with open(filename, "r", encoding="utf-8") as file:
            content = file.read()
            return content
    except FileNotFoundError:
        print(f"File '{filename}' not found.")
    finally:
        print("completed.")



print("===== Test 1: python_notes.txt =====")
result = read_file_safe("python_notes.txt")
if result:
    print(result)


print("===== Test 2: No_file.txt =====")
result = read_file_safe("No_file.txt")
```

```python
def safe_get(url):
    try:
        response = requests.get(url, timeout=5)
        return response
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print(f"Unexpected error: {e}")
    return None 



def safe_post(url, payload):
    try:
        response = requests.post(url, json=payload, timeout=5)
        return response
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print(f"Unexpected error: {e}")
    return None 


    
response = safe_get("https://dummyjson.com/products?limit=20")

if response and response.status_code == 200:
    products = response.json()["products"]
    print(f"Fetched {len(products)} products successfully!\n")

    # Filter by rating >= 4.5
    top_rated = [p for p in products if p["rating"] >= 4.5]

    # Sort by price descending
    sorted_products = sorted(top_rated, key=lambda p: p["price"], reverse=True)
    for p in sorted_products:
        print(f"{p['id']:<4} | {p['title'][:30]:<30} | {p['category'][:13]:<13} | ${p['price']:<7} | {p['rating']}")
elif response:
    print(f"HTTP Error: {response.status_code}") 

    print("\n===== Step 3: Fetch Laptops =====\n")

response = safe_get("https://dummyjson.com/products/category/laptops")

if response and response.status_code == 200:
    laptops = response.json()["products"]
    print("\n")
    print(f"Fetched {len(laptops)} laptops successfully!\n")

    for laptop in laptops:
        print(f"{laptop['title']:<35} | ${laptop['price']}")
elif response:
    print(f"HTTP Error: {response.status_code}")


    print("\n===== Step 4: POST New Product =====")
new_product = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}

response = safe_post("https://dummyjson.com/products/add", new_product)

if response and response.status_code == 201:
    print("\n")
    print("Product created successfully!\n")
    print(json.dumps(response.json(), indent=4))
elif response:
    print(f"HTTP Error: {response.status_code}")
```
